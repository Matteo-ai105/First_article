```python
import pandas as pd
import numpy as np
import h2o 
from h2o.estimators.gbm import H2OGradientBoostingEstimator
h2o.init()
```

Upload the Simulated xlsx database

```python
Sample = pd.read_excel('Simulation Sample.xlsx')
Sample = h2o.H2OFrame(Sample)
print(Sample.head)

response = 'ID'
Sample[response] = Sample[response].asfactor()
predictors = Sample.columns
del predictors[0:2]
print(predictors)
```

Pre-processing phase: Splitting the dataset in training and testing

```python
train,test = Sample.split_frame(ratios=[.8], 
                                      seed = 1234,
                                     destination_frames=['train.hex', 'test.hex'])
print(train.describe())
print(test.describe())
```

```python
cv_gbm = H2OGradientBoostingEstimator(ntrees = 1000, nfolds = 10, seed = 0xDECAF)
cv_gbm.train(x = predictors, y = response, training_frame = train)
```

```python
perf_cv = cv_gbm.model_performance(xval=True)
```

```python
print(cv_gbm.cross_validation_metrics_summary())
print(perf_cv)
print(cv_gbm)
```

```python
preds = cv_gbm.predict(test)
preds.head()
```

```python
print(cv_gbm.model_performance(test))
```

```python
New_Sample = pd.read_excel('Standardization_NDR.xlsx')
New_Sample = h2o.H2OFrame(New_Sample)
print(New_Sample.head)
```
